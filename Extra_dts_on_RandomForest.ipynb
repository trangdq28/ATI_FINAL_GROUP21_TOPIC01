# Random Forest base line

```python
# 1. IMPORT & LOAD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import joblib

# Load dataset
DATA_PATH = "data.csv"
df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
print(df.head())

# 2. MISSING VALUES & CLEANING
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Fill numeric with median
for c in num_cols:
    if df[c].isna().sum() > 0:
        df[c].fillna(df[c].median(), inplace=True)

print("Missing after fill:", df.isna().sum().sum())

# 3. TARGET PREPARATION
TARGET = "Diabetes_binary"

# Fix labels: chuyển giá trị >1 thành 1
print("Unique labels before fix:", sorted(df[TARGET].unique()))
df[TARGET] = df[TARGET].apply(lambda x: 1 if x > 1 else int(x))
print("Unique labels after fix:", sorted(df[TARGET].unique()))

# Target distribution
counts = df[TARGET].value_counts()
ratio = df[TARGET].value_counts(normalize=True) * 100
print("\nTarget Distribution:")
print("Counts:\n", counts)
print("Ratio (%):\n", ratio.round(2))

# 4. SPLIT DATA
X = df.drop(columns=[TARGET])
y = df[TARGET].astype(int)

# Save feature columns
feature_columns = X.columns.tolist()
print("\nFeature count:", len(feature_columns))

# Train-test split (stratify theo nhãn)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train label distribution:\n", y_train.value_counts(normalize=True))

# 5. TRAIN RF WITH BALANCED CLASS WEIGHTS
print("\n" + "="*60)
print("TRAINING RF WITH BALANCED CLASS WEIGHTS")
print("="*60)

rf_balanced = RandomForestClassifier(
    n_estimators=200,
    class_weight='balanced',  # Tự động cân bằng class weights
    random_state=42,
    n_jobs=-1
)

rf_balanced.fit(X_train, y_train)
print("✓ Training complete!")

# 6. EVALUATION
y_pred = rf_balanced.predict(X_test)
y_proba = rf_balanced.predict_proba(X_test)[:,1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
roc_auc = roc_auc_score(y_test, y_proba)
cm = confusion_matrix(y_test, y_pred)

print("\n" + "="*60)
print("EVALUATION RESULTS")
print("="*60)
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
print(f"ROC-AUC  : {roc_auc:.4f}")
print(f"\nConfusion Matrix:")
print(cm)

# Classification Report
print("\n" + "="*60)
print("CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_test, y_pred, target_names=['No Diabetes', 'Diabetes']))

# 7. VISUALIZATIONS

# Confusion Matrix Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.title('Confusion Matrix - RF Balanced', fontsize=14)
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.show()

# Training Progress: OOB Score per tree
print("\n" + "="*60)
print("TRAINING PROGRESS - OOB SCORE")
print("="*60)

# Train với oob_score để track training progress
rf_balanced_oob = RandomForestClassifier(
    n_estimators=200,
    class_weight='balanced',
    oob_score=True,
    warm_start=True,
    random_state=42,
    n_jobs=-1
)

oob_scores = []
n_trees = [10, 20, 30, 50, 75, 100, 125, 150, 175, 200]

for n in n_trees:
    rf_balanced_oob.n_estimators = n
    rf_balanced_oob.fit(X_train, y_train)
    oob_scores.append(rf_balanced_oob.oob_score_)
    print(f"Trees: {n:3d} | OOB Score: {rf_balanced_oob.oob_score_:.4f}")

# Plot OOB Score progression
plt.figure(figsize=(10, 6))
plt.plot(n_trees, oob_scores, marker='o', linewidth=2, markersize=8, color='steelblue')
plt.xlabel('Number of Trees', fontsize=12)
plt.ylabel('OOB Score', fontsize=12)
plt.title('Random Forest Training Progress (OOB Score)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ROC Curve
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_proba)

plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, linewidth=2, label=f'RF Balanced (AUC = {roc_auc:.3f})', color='steelblue')
plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve - RF Balanced', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Precision-Recall Curve
from sklearn.metrics import precision_recall_curve, average_precision_score

precisions, recalls, pr_thresholds = precision_recall_curve(y_test, y_proba)
avg_precision = average_precision_score(y_test, y_proba)

plt.figure(figsize=(10, 6))
plt.plot(recalls, precisions, linewidth=2, label=f'RF Balanced (AP = {avg_precision:.3f})', color='steelblue')
plt.xlabel('Recall', fontsize=12)
plt.ylabel('Precision', fontsize=12)
plt.title('Precision-Recall Curve - RF Balanced', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Learning Curve (Train vs Validation Score)
from sklearn.model_selection import learning_curve

print("\n" + "="*60)
print("LEARNING CURVE - COMPUTING...")
print("="*60)

train_sizes, train_scores, val_scores = learning_curve(
    rf_balanced,
    X_train, y_train,
    cv=5,
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='f1',
    random_state=42
)

train_mean = train_scores.mean(axis=1)
train_std = train_scores.std(axis=1)
val_mean = val_scores.mean(axis=1)
val_std = val_scores.std(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', linewidth=2, label='Training F1', color='steelblue')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2, color='steelblue')
plt.plot(train_sizes, val_mean, 'o-', linewidth=2, label='Validation F1', color='coral')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.2, color='coral')
plt.xlabel('Training Set Size', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.title('Learning Curve - RF Balanced', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Metrics Comparison Bar Chart
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
metrics_values = [accuracy, precision, recall, f1, roc_auc]

plt.figure(figsize=(10, 6))
bars = plt.bar(metrics_names, metrics_values, color=['steelblue', 'coral', 'lightgreen', 'gold', 'plum'])
plt.ylabel('Score', fontsize=12)
plt.title('Performance Metrics - RF Balanced', fontsize=14)
plt.ylim(0, 1)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# Feature Importances (Top 15)
feat_imp = pd.Series(rf_balanced.feature_importances_, index=feature_columns).sort_values(ascending=False)

print("\n" + "="*60)
print("TOP 15 FEATURE IMPORTANCES")
print("="*60)
print(feat_imp.head(15))

plt.figure(figsize=(10, 8))
feat_imp.head(15).sort_values().plot(kind='barh', color='steelblue')
plt.xlabel('Importance', fontsize=12)
plt.title('Top 15 Feature Importances (RF Balanced)', fontsize=14)
plt.tight_layout()
plt.show()

# 8. SAVE MODEL
print("\n" + "="*60)
print("SAVING MODEL")
print("="*60)

joblib.dump(rf_balanced, "rf_balanced_model.joblib")
pd.Series(feature_columns).to_csv("feature_columns.csv", index=False, header=False)

print("✓ Model saved as: rf_balanced_model.joblib")
print("✓ Feature columns saved as: feature_columns.csv")
print("\n✓ Done!")
```
